```python
def plot_correlation_map( df ):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(corr, cmap = cmap,square=True, cbar_kws={ 'shrink' : .9 }, ax=ax, annot = True, annot_kws = { 'fontsize' : 12 })
plot_correlation_map(dp)
```

This function gives us a heat map showing us the correlation between different variables, 
and for our target which is "Survived", we can interpet that the most important features in a decreasing order are :

"Sex" (the more Survived, the more "Sex" goes down, and female is represented as 0, so females are more likely to survive )

"Pclass" (negative correlation, so the more survived, the more "Pclass" goes down, so the higher the classe is,(1,2,3), the more the passenger is likely to survive)

"Fare"

"Embarked"

"Parch"

"Age", 

"SibSp"

"PassengerId"

```python
def title(x):
    if ',' in x and '.' in x :
        return x[x.index(',')+2:x.index('.')]
dp['Title']=dp["Name"]
dp['Title']=dp["Title"].apply(title)
dp
```

```python
pd.crosstab(dp.Title,dp.Sex).plot(kind='bar',figsize=(10,10))
plt.legend(['Female', 'Male'])
```

```python
dp.boxplot(column = 'Age', by = 'Title',figsize=(8,8));
plt.title('')
plt.xticks(rotation=-75)
```

```python
dp.boxplot(column = 'Fare', by = 'Title',figsize=(10,10));
plt.title('')
plt.xticks(rotation=-75)
```
