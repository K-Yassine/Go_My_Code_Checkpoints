```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
       	['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
       	['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
       	['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
        ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary,columns=te.columns_)
```

```python
plt.figure(figsize=(9,6))
liste=[]
for loop in df.columns:
    liste.append(df[loop].value_counts()[1])
plt.bar(df.columns,liste)
plt.xticks(rotation=-75)
```

```python
data = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
       	['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
       	['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
       	['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
        ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]

# Pull out combinations
from itertools import combinations
data_pairs = []
for d in data:
    data_pairs += [list(sorted(x)) + [1] for x in combinations(d, 2)]
    # Add reverse as well (this will mirror the heatmap)
    data_pairs += [list(sorted(x))[::-1] + [1] for x in combinations(d, 2)]

dfr = pd.DataFrame(data_pairs)
dfr_zeros = pd.DataFrame([list(x) + [0] for x in combinations(dfr[[0, 1]].values.flatten(), 2)])
dfr = pd.concat((dfr, dfr_zeros))
dfr = dfr.groupby([0, 1])[2].sum().reset_index().pivot(0, 1, 2).fillna(0)


sns.heatmap(dfr, cmap='YlGnBu')
plt.show()
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df,min_support=0.4)
```

```python
frequent_itemsets = apriori(df,min_support=0.6,use_colnames=True )
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
```

```python
rules = association_rules(frequent_itemsets,metric="lift")
rules
```

```python
sns.set(rc={'figure.figsize':(20,13)})
axes = rules.plot.bar(rot=0, subplots=True)
axes[1].legend(loc=2);
```

### Changing our data frame

```python
liste_1 = data.values.tolist()
```

```python
liste_2 = []
for i in liste_1:
    x = []
    for j in i:
        if str(j) !='nan':
            x.append(j)
    liste_2.append(x)   
```

```python
te1 = TransactionEncoder()
te_ary1 = te1.fit(liste_2).transform(liste_2)
df1 = pd.DataFrame(te_ary1,columns = te1.columns_)
```

```python
df1
```

```python
frequent_itemsets1 = apriori(df1,min_support=0.01,use_colnames=True )
frequent_itemsets1.head()
```

```python
association_rules(frequent_itemsets1,metric="confidence",min_threshold=0.2).head()
```

```python
rules1 = association_rules(frequent_itemsets1,metric="lift",min_threshold=2)
rules1.head()
```

```python
sns.set(rc={'figure.figsize':(25,15)})
axes = rules1.plot.bar(rot=0, subplots=True)
axes[1].legend(loc=2);
```
